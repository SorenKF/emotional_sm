```python
import pandas as pd, numpy as np
from pathlib import Path
```

```python

def convert_file(file_path, dataset_type=None):
    """Convert a file in either CSV or TSV format to the VUA format and save as CSV.

    Args:
        file_path (str): a path to a data file.
    Optional:
        dataset_type (str): a string indicating the type of the dataset scheme (e.g. "OLID")
    """    
    
    vua_header = ['id' , 'text' , 'label']
    
    

    # check the dataset input type (TRAC/MLMA/OLID)
    if dataset_type == 'TRAC':
        dataset_converted = pd.read_csv(file_path , names=vua_header)
    if dataset_type == 'mlma':
        dataset_converted = pd.read_csv(file_path , sep='\t' , names=vua_header) # the header was manually edited ahead of this.
    if dataset_type == "olid":
        dataset_converted = pd.read_csv(file_path , sep='\t' , names=vua_header)

    
    return dataset_converted

```

```python
# Test on TRAC

output = convert_file("../Data/OLID_full/converted/levelc/test.tsv",  dataset_type='olid')

# Write TRAC files
output.to_csv('../Data/OLID_full/converted/levelc/test.tsv' , sep='\t', index=False)
```
